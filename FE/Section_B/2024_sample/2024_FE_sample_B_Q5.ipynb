> 注文データから，商品 $x$ と商品 $y$ とが同一の注文で購入されやすい傾向を示す関連度 $L_{xy}$ を，次の式で計算する。

$$
L_{xy} = \frac{M_{xy} \times 全注文数　　　}{K_x \times K_y}
$$

> ここで，$M_{xy} $ は商品 $x$ と商品 $y$ とが同一の注文で購入された注文数，$K_{x}$ は商品 $x$ が購入された注文数，$K_y$ は商品 $y$ が購入された注文数を表す。表の例では，$M_{AB}$ が $2$，全注文数が $6$，$K_A$ が $4$，$K_B$ が $3$ であるので，商品 A と商品 B の関連度 $L_{AB}$ は，$(2 \times 6) \div (4 \times 3) = 1.0$である。
>
> 手続 putRelatedItem は，大域変数 orders に格納された注文データを基に，引数で
与えられた商品との関連度が最も大きい商品のうちの一つと，その関連度を出力する。
プログラムでは，商品は文字列で表し，注文は購入された商品の配列，注文データは
注文の配列で表している。注文データには 2 種類以上の商品が含まれるものとする。
また，注文データにある商品以外の商品が，引数として与えられることはないものと

**正解**
- オ　a: arrayM[i], b: arrayK[i], c: ordersの要素数

**解説**

まず** ［　c　］ **に入るものは，$L_{xy}$の計算式の定義の通り，全注文数と同値になるordersの要素数になります。

$M_{xy}$は注文時に$x$と$y$が同時に注文された数，$K_{x}$は$x$の注文数で，$M_{xy}$ は，配列 `arrayM`，$K_{x}$は，配列 `arrayK`にそれぞれ格納されます。

関数の引数`item`には計算対象となる注文商品が入っているので，** ［　a　］ **では注文の中に`item`と等しいものがあるとき真になるとき実行されます。従って，$M_{xy}$の計算になります。

** ［　b　］ **は，商品ごとの注文数をカウントする処理が実行されます。従って，$M_{x}$の計算になります。

### Pythonでの実装例

````python
orders = [["A", "B", "D"], ["A", "D"], ["A"], ["A", "B", "E"], ["B"], ["C", "E"]]


def putRelatedItem(item):
  # 2次元リストのordersを （フラット化）
  dummyItems = [item for sublist in orders for item in sublist]
  allItems = list(set(dummyItems))
  otherItems = list(set(dummyItems) - set(item))

  itemCount = 0
  arrayK = [0] * len(otherItems)
  arrayM = [0] * len(otherItems)
  valueL, maxL = 0, -float("inf")
  relatedItem = ""

  for order in orders:
    if item in order:
      itemCount += 1
    for i in range(1, len(otherItems)+1, 1):
      if otherItems[i-1] in order:
        if item in order:
          arrayM[i-1] += 1
        arrayK[i-1] += 1
  for i in range(1, len(otherItems) + 1, 1):
    valueL = (arrayM[i-1] * len(orders)) / (itemCount * arrayK[i-1])
    if valueL > maxL:
      maxL = valueL
      relatedItem = otherItems[i-1]
  print(relatedItem, maxL)

# item を A として putRelatedItemを呼び出す
putRelatedItem('A')
````
